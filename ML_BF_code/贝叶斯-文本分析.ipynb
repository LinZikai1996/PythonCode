```python
import pandas as pd
import jieba
# pip install -i https://pypi.tuna.tsinghua.edu.cn/simple jieba
# 数据源：http://www.sogou.com/labs/resource/ca.php
%matplotlib notebook
```

```python
new_dataFrame = pd.read_table('E:\\project\\python\\machineLearning\\ML_BeiFeng\\data\\sogouData\\val.txt',names=['category','theme','URL','content'],encoding='utf-8')
new_dataFrame = new_dataFrame.dropna()
new_dataFrame.head()
```

```python
new_dataFrame.shape
```

```python
content = new_dataFrame["content"].tolist()
print(content[1000])
```

```python
content_s = []

for line in content:
    current_segment = jieba.lcut(line)
    if len(current_segment) > 1 and current_segment != '\r\n':
        content_s.append(current_segment)
```

```python
content_s[1000]
```

```python
content_dataFrame = pd.DataFrame({'content_s':content_s})
content_dataFrame.head()
```

```python
# 停词表
stopwords=pd.read_csv("E:\project\python\machineLearning\ML_BeiFeng\data\sogouData\stopwords.txt",index_col=False,sep="\t",quoting=3,names=['stopword'], encoding='utf-8')
stopwords.head(20)
```

```python
def drop_stop_words(contents, stopwords):
    content_clean = []
    all_words = []
    for line in contents:
        line_clean = []
        for word in line:
            if word in stopwords:
                continue
            line_clean.append(word)
            all_words.append(str(word))
        content_clean.append(line_clean)
    return content_clean, all_words

contents_clean, all_words = drop_stop_words(content_dataFrame["content_s"].tolist(), stopwords["stopword"].tolist())
```

```python
content_dataFrame = pd.DataFrame({"contents_clean":contents_clean})
content_dataFrame.head()
```

```python
all_words_dataFrame = pd.DataFrame({'all_words':all_words})
all_words_dataFrame.head()
```

```python
# 统计词频
words_count = all_words_dataFrame.groupby(['all_words']).size()\
    .reset_index(name='count').sort_values(by=["count"],ascending=False)
words_count.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0, 5.0)

wordcloud=WordCloud(font_path="E:\project\python\machineLearning\ML_BeiFeng\data\sogouData\simhei.ttf",background_color="white",max_font_size=80)
word_frequence = {x[0]:x[1] for x in words_count.head(100).values}
wordcloud=wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```
